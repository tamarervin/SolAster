# Looking at the outlier magnetic observables: Dec 30, 2020

```python
import numpy as np
import pandas as pd
import tamar.tools.plotting_funcs as plot
```

Read in velocities from csv.

```python
# csv file with rv components
csv_file = '/Users/tervin/csv_files/neid/12_30_20.csv'
```

```python
# create pandas dataframe
component_df = pd.read_csv(csv_file)

# get dates
dates = component_df.date_obs.values

# get velocities lists
v_phot = component_df.v_phot.values
v_conv = component_df.v_conv.values

# get magnetic observables
f = component_df.f.values
Bobs = component_df.Bobs.values
f_bright = component_df.f_bright.values
f_spot = component_df.f_spot.values
```

Get days range for plotting.

```python
# get range based on good indices
days_range = range(0, len(Bobs))
```

## Plotting of Bobs and filling factor

### Timeseries plot

```python
x = days_range
y_list = [f, Bobs]
xlabel = ['Observations']
ylabel_list = ['Filling Factor (%)', 'Unsigned Flux (G)']
title = 'Comparison between Unsigned Flux and Filling Factor'

plot.vert_comp_timeseries(x, y_list, title, xlabel, ylabel_list, save_fig=None)
```

### Correlation plot

```python
x = f
y = Bobs
xlabel = 'Filling Factor (%)'
ylabel = 'Unsigned Flux (G)'
title = 'Correlation between Unsigned Flux and Filling Factor'

plot.correlation_plot(x, y, title, xlabel, ylabel, save_fig=None)
```

## Get outlier dates

```python
outlier_inds = np.logical_and(Bobs > 20, f > 10)
outlier_dates = dates[outlier_inds]
```

```python
print(outlier_dates)
```

## Same plots without outliers for comparison

```python
good_f = f[~outlier_inds]
good_Bobs = Bobs[~outlier_inds]
good_days = range(0, len(good_Bobs))
```

### Timeseries plot

```python
x = days_range
y_list = [good_f, good_Bobs]
xlabel = ['Observations']
ylabel_list = ['Filling Factor (%)', 'Unsigned Flux (G)']
title = 'Comparison between Unsigned Flux and Filling Factor (Outliers Removed)'

plot.vert_comp_timeseries(x, y_list, title, xlabel, ylabel_list, save_fig=None)
```

### Correlation plot

```python
x = good_f
y = good_Bobs
xlabel = 'Filling Factor (%)'
ylabel = 'Unsigned Flux (G)'
title = 'Correlation between Unsigned Flux and Filling Factor (Outliers Removed)'

plot.correlation_plot(x, y, title, xlabel, ylabel, save_fig=None)
```
