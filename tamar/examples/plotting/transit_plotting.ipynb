```python
import os
import numpy as np
import pandas as pd
import tamar.tools.plotting_funcs as plot
import matplotlib.pyplot as plt
from tamar.tools.settings import CsvDir, ImgDir
```

Read in velocities from csv.

```python
# csv file with rv components
csv_file = os.path.join(CsvDir.MERCURY, '2016_sunspot.csv')
```

Get component lists.

```python
# create pandas dataframe
component_df = pd.read_csv(csv_file)
```

```python
# csv file with rv components
# csv_file = os.path.join(CsvDir.MERCURY, '2019_2min.csv')

# create pandas dataframe
component_df = pd.read_csv(csv_file)
# get velocities lists
v_phot = component_df.v_phot.values
vphot_bright = component_df.vphot_bright.values
vphot_spot = component_df.vphot_dark.values
v_conv = component_df.v_conv.values
# rv_sun = component_df.rv_sun.values
rv_model = component_df.rv_model.values

# get magnetic observables
f = component_df.f.values
Bobs = component_df.Bobs.values
f_bright = component_df.f_bright.values
f_spot = component_df.f_spot.values

# get dates list
dates = component_df.date_obs.values
date_jd = component_df.date_jd.values
```

## Plotting magnetic stuff

```python
save_fig = None
```

```python
x = date_jd
y_list = [f, Bobs]
xlabel = 'JD Date'
ylabel_list = ['Filling Factor (%)', 'Unsigned Flux (G)']
title = 'Comparison between Unsigned Flux and Filling Factor'

plot.vert_comp_timeseries(x, y_list, title, xlabel, ylabel_list, save_fig)
```

```python
x = f
y = Bobs
xlabel = 'Filling Factor (%)'
ylabel = 'Unsigned Flux (G)'
title = 'Correlation between Unsigned Flux and Filling Factor'
# save_fig = '/Users/tervin/images/neid/correlation_f_Bobs.png'

plot.correlation_plot(x, y, title, xlabel, ylabel, save_fig)
```

Comparing observables to RV model.

```python
x_list = [f, Bobs]
y = rv_model
title = 'Correlation between observables and model RV variation'
xlabel_list = ['Filling Factor (%)', 'Observed Unsigned Flux (G)']
ylabel = 'Model RV Variation'
# save_fig = '/Users/tervin/images/neid/correlation_rvmodel_f_Bobs.png'

plot.two_correlation_plots(x_list, y, title, xlabel_list, ylabel, save_fig)
```

```python
x = date_jd
y_list = [Bobs, v_conv, v_phot, rv_model]
xlabel = 'JD Date'
ylabel_list = [r'$\rm B_{\rm obs}\/\rm (G)$', r'$\rm v_{\rm conv}\/\rm (m/s)$', r'$\rm v_{\rm phot}\/\rm (m/s)$', r'$\rm rv_{\rm model}\/\rm(m/s)$']
title = 'Mercury 2019 Transit'

save_fig=None
# save_fig = os.path.join(ImgDir.MERCURY_IMG, '2019_mercury_transit.png')
plot.vert_comp_timeseries(x, y_list, title, xlabel, ylabel_list, save_fig)
save_fig = None
print(np.std(rv_model))
```
