# Comparing to NEID RVs

```python
# fit a line to the neid and model RVs
from sklearn.linear_model import LinearRegression

# dates and rvs
use = np.where(date_jd == neid_date_jd[0])[0][0]
x = (date_jd[use:] - date_jd[use]) * 24 * 60
x2 = (neid_date_jd - neid_date_jd[0]) * 24 * 60
rv_sun_plot = rv_sun - np.median(rv_sun)
rv_model_plot = rv_model[use:] - np.median(rv_model)

# fit NEID RV
reg = LinearRegression().fit(x2.reshape(-1, 1), rv_sun_plot)
sun_fit = reg.predict(x2.reshape(-1, 1))

# fit model RV
mreg = LinearRegression().fit(x.reshape(-1, 1), rv_model_plot)
model_fit = mreg.predict(x.reshape(-1, 1))
```

```python
y_list = [Bobs[use:], rv_sun_plot, rv_model_plot]
xlabel = 'Minutes'
ylabel_list = [r'$\rm B_{\rm obs}\/\rm [m/s]$', r'$\rm rv_{\rm sun}\/\rm [m/s]$',
               r'$\rm rv_{\rm model}\/\rm [m/s]$']

plot_style = os.path.join(PlotDir.MPL, 'figure_series.mplstyle')
plt.rcParams['mathtext.fontset'] = 'custom'
plt.rcParams['mathtext.cal'] = 'Helvetica Neue LT Pro'
# plt.rcParams['axes.unicode_minus'] = True
plt.style.use(plot_style)

# set up figure
fig, axs = plt.subplots(len(y_list), 1, sharex='all', figsize=[12, 2 * len(y_list)],  gridspec_kw={'hspace': 0})

# set up axes labels
for i in range(0, len(axs)):
    axs[i].set(ylabel=ylabel_list[i])
axs[2].set(xlabel=xlabel)
axs[0].set(ylim=(6.8, 9.2), yticks=np.arange(7, 9.2, step=0.5))
axs[1].set(ylim=(-4.2, 4.2), yticks=np.arange(-4, 4.52, step=2))
axs[2].set(ylim=(-1.2, 1.2), yticks=np.arange(-1, 1.2, step=0.5))

# plot data
color_list = ['lightcoral', 'lightblue', 'plum', 'pink', 'cornflowerblue']
axs[0].scatter(x, y_list[0], color=color_list[0], edgecolors='k', linewidths=1.0)
axs[1].scatter(x2, y_list[1], color=color_list[1], edgecolors='k', linewidths=1.0)
axs[2].scatter(x, y_list[2], color=color_list[2], edgecolors='k', linewidths=1.0)

# plot fit lines
axs[1].plot(x2, sun_fit, color='k', linewidth=2, label='Slope: ' + str(np.round(reg.coef_[0], 4)))
axs[2].plot(x, model_fit, color='k', linewidth=2, label='Slope: ' + str(np.round(mreg.coef_[0], 4)))

for i in [1, 2]:
    leg = axs[i].legend(handlelength=0, handletextpad=0, loc='upper right')
    for item in leg.legendHandles:
        item.set_visible(False)

# align y axis labels
fig.align_ylabels(axs)
fig.suptitle('2021/02/07')

# print scatter
print(np.std(rv_sun_plot), np.std(rv_model_plot))
```

# Plot Periodogram

```python
from scipy.signal import lombscargle
```

```python
rv = rv_model
jd = date_jd
```

```python
# rough time sampling
djd = np.median(np.diff(jd)) # days
djd_s = djd * 24. * 3600. # seconds
djd_hz = 1. / djd_s # hz
freqs = np.linspace(djd_hz / 4, djd_hz/3., 3500)
drv = (rv - np.mean(rv))*1e3
jd_s = (jd - min(jd)) * 24. * 3600.
ls = lombscargle(jd_s, drv, freqs)
def tick_function(X):
    V = (1./ (X * 1e-3)) * 1/60.
    return ['%.3f' % z for z in V]
fig = plt.figure()
ax1 = fig.add_subplot(111)
ax2 = ax1.twiny()
xvals = freqs * 1e3
data = ls
ax1.plot(xvals, data, 'k',lw=2,alpha=0)
ax1.plot(xvals, data, 'r',lw=0.5,alpha=0)
ax1Ticks = ax1.get_xticks()
ax2Ticks = ax1Ticks
ax2.set_xticks(ax2Ticks)
ax2.set_xbound(ax1.get_xbound())
ax2.set_xticklabels(tick_function(ax2Ticks))
ax1.set_xlabel('Frequency [mHz]')
ax2.set_xlabel('Period [min]', labelpad=10)
ax1.set_ylabel('LS power')
ax1.grid(True)
ax2.plot(xvals, data, 'black', lw=4.)
ax2.plot(xvals, data, 'orchid', lw=2.)
title = ax1.set_title('Solar RV Periodogram')
title.set_y(1.2)
plt.show()


```
