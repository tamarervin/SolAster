```python
import numpy as np
import pandas as pd
import tamar.tools.plotting_funcs as plot
import matplotlib.pyplot as plt
from scipy import optimize
```

Read in velocities from csv.

```python
# csv file with rv components
csv_file = '/Users/tervin/csv_files/neid/calcs/rvs_from_txt.csv'
```

```python
# create pandas dataframe
df = pd.read_csv(csv_file)

# sort dates
jd_dates = df.date_jd.values
inds = np.argsort(jd_dates)
jd_dates = jd_dates[inds]

# get data
dates = df.date_obs.values[inds]
v_conv = df.v_conv[inds]
v_quiet = df.v_quiet[inds]
rv_sun = df.rv_sun[inds]
rv_error = df.rv_error[inds]
f = df.f.values[inds]
Bobs = df.Bobs.values[inds]
f_bright = df.f_bright.values[inds]
f_spot = df.f_spot.values[inds]
f_small = df.f_small.values[inds]
f_large = df.f_large.values[inds]
f_network = df.f_network.values[inds]
f_plage = df.f_plage.values[inds]
f_nonconv = df.f_nonconv.values[inds]
flux_quiet = df.quiet_flux.values[inds]
flux_ar = df.ar_flux.values[inds]
flux_conv = df.conv_flux.values[inds]
flux_pol = df.pol_flux.values[inds]
flux_pol_conv = df.pol_conv_flux.values[inds]
vconv_quiet = df.vconv_quiet.values[inds]
vconv_large = df.vconv_large.values[inds]
vconv_small = df.vconv_small.values[inds]
```

## Plotting stuff
